```python
import pandas as pd
import re
# file: data_demHawaii.json
df = pd.read_csv('https://raw.githubusercontent.com/kathanner/hcan/main/kids_count_data.csv')
df = df.replace('NaN', pd.NA).dropna(subset=["StatNow","StatPrevious"])
# filtering the rows
lHawaii = ["Hawaii"]
iDemographics = ['Population of young children','Percent of families with children','Percent of families with children that are single-parent families','Percent of families with young children','Percent of families with children that have young children','Percent of children in immigrant families']
df = df[df['Location'].isin(lHawaii)]
df = df[df['Indicator'].isin(iDemographics)]
df_demographics = df[["Indicator","DataType","StatNow","StatPrevious","StatIncrease","TimeFrame","TimeFramePrevious","Source"]].to_dict('records')
df['demographicsHawaii'] = df_demographics
df_demographics = re.sub(r'\\', '', df[['demographicsHawaii']].to_json(orient='records', indent=4))
print(df_demographics)
```
