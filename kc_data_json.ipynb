```python
import pandas as pd
import re
# file: data_demHawaii.json
df = pd.read_csv('https://raw.githubusercontent.com/kathanner/hcan/main/kids_count_data.csv')
lHawaii = ["Hawaii"]
category = ['Demographics']
df = df[df['Location'].isin(lHawaii)]
df = df[df['Category'].isin(category)]
df_demographics = df[["Indicator","CurrentYear","PreviousYear","Current","Previous","Change","Source"]].to_dict('records')
df['demographicsHawaii'] = df_demographics
df_demographics = re.sub(r'\\', '', df[['demographicsHawaii']].to_json(orient='records', indent=4))
print(df_demographics)
```
